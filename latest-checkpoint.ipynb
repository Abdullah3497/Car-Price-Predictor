```python
import pandas as pd
from xgboost import XGBRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score

# üì• Load data
data = pd.read_csv("car_listings.csv")

# üßº Clean 'price'
data['price'] = (
    data['price']
    .astype(str)
    .str.replace(',', '', regex=False)
    .str.extract(r'(\d+\.?\d*)')[0]
    .astype(float)
)

# üßº Clean other numeric columns
data['mileage'] = data['mileage'].astype(str).str.replace(',', '').str.extract(r'(\d+)').astype(float)
data['engine'] = data['engine'].astype(str).str.replace(',', '').str.extract(r'(\d+)').astype(float)
data['year'] = pd.to_numeric(data['year'], errors='coerce')

# ‚ùå Drop 'url' if exists
data.drop(columns=['url'], inplace=True, errors='ignore')

# üÜï Extract 'make' and 'model' from 'title'
data['make'] = data['title'].str.split().str[0]
data['model'] = data['title'].str.split().str[1]

# ‚ùå Drop rows with missing essential data
data.dropna(subset=['price', 'mileage', 'engine', 'year', 'make', 'model'], inplace=True)

# üéØ Define features and target
X = data[['category', 'make', 'model', 'city', 'year', 'mileage', 'fuel', 'engine', 'transmission']]
y = data['price']

# üîÑ One-hot encoding for categorical features
X_encoded = pd.get_dummies(X, drop_first=True)

# ‚úÖ Save column names for prediction alignment
columns_for_model = X_encoded.columns

print(data.head())

# ‚úÇÔ∏è Split data
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42)

# üöÄ Train model
model = XGBRegressor(n_estimators=100, learning_rate=0.1, random_state=42)
model.fit(X_train, y_train)

# üìä Evaluate
y_pred = model.predict(X_test)
print(f"üìà R¬≤ Score: {r2_score(y_test, y_pred):.4f}")

import pickle

# ‚úÖ Save model
with open("car_price_model.pkl", "wb") as f:
    pickle.dump(model, f)

# ‚úÖ Save feature columns for alignment during prediction
with open("model_columns.pkl", "wb") as f:
    pickle.dump(columns_for_model.tolist(), f)
```

```python
# üì• User input
user_input = {
    'category': input("Enter category (e.g., 5 Seater): "),
    'make': input("Enter car brand (e.g., Toyota): "),
    'model': input("Enter car model (e.g., Corolla): "),
    'city': input("Enter city (e.g., Lahore): "),
    'year': int(input("Enter year (e.g., 2021): ")),
    'mileage': float(input("Enter mileage in KM (e.g., 35000): ")),
    'fuel': input("Enter fuel type (e.g., Petrol): "),
    'engine': float(input("Enter engine capacity in CC (e.g., 1800): ")),
    'transmission': input("Enter transmission (e.g., Automatic): ")
}

# üîÑ Convert to DataFrame and encode
user_df = pd.DataFrame([user_input])
user_encoded = pd.get_dummies(user_df)
user_encoded = user_encoded.reindex(columns=columns_for_model, fill_value=0)

# üìà Predict
predicted_price = model.predict(user_encoded)[0]
print(f"\nüí∞ Predicted Car Price: Rs. {predicted_price:,.0f}")



```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 6))
sns.scatterplot(x=y_test, y=y_pred)
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title("Actual vs Predicted Car Prices")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
importances = model.feature_importances_
feature_names = X_train.columns

# Top 20 features (optional)
top_n = 20
sorted_idx = importances.argsort()[::-1][:top_n]

plt.figure(figsize=(10, 6))
sns.barplot(
    x=importances[sorted_idx],
    y=feature_names[sorted_idx],
    palette="viridis"
)
plt.xlabel("Feature Importance Score")
plt.title("Top Feature Importances (XGBoost)")
plt.tight_layout()
plt.show()
```
